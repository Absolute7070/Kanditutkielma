```python
def gradient_atApoint(function, values: list,  h: float = 0.001): 
    gradient = [twoPointDerivative(function = function, h=h, values=values, whichParameter=i) for i in range(len(values))         ]
    return gradient 
```

```python
def twoPointDerivative(function, h: float, values: list, whichParameter: int) -> float:
    values_minusH = np.array(values).copy()
    values_plusH = np.array(values).copy()
    
    values_minusH[whichParameter] -= h 
    values_plusH[whichParameter] += h 
    
    numerator = (function(values_plusH) - function(values_minusH)) 
    denominator = 2 * h 
    
    return numerator/denominator
```
