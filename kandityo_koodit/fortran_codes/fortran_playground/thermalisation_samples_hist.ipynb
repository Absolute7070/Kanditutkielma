# testing what happens to thermalisation_samples when changing coordinate range 

### 1.samples, coordinate range [-3, 3]

```python
df = open("thermalisation_samples.txt", 'r')
df_content  = df.read()
df_content_as_rows = df_content.split("\n")

df_content_as_rows[:5]
```

```python
# convert data into numbers 

data = []

for row_index in range(len(df_content_as_rows)): 
    data.append(list(map(lambda x: float(x), df_content_as_rows[row_index].split())))

```

```python
data.pop(-1)
data = np.array(data).flatten()
```

```python
len(data)
```

```python
plt.hist(data)
```
