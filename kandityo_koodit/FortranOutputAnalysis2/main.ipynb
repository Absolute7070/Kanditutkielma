```python
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv('run2_2_gradient', names = ['A', 'B'])
fig, axes = plt.subplots(nrows= 1, ncols=2, figsize = (10,5))
axes[0].scatter(x= df.index, y =  df.A)
axes[1].scatter(x= df.index, y =  df.B)
fig.suptitle('params', fontsize = 16 )
```
